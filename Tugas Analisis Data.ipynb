## TUGAS ANALISA DATA
Analisa data UMKM tahun 1997-2012 di Indonesia



```python
import pandas as pd

csv = pd.read_csv('processed-data-umkm-1997-2012.csv', delimiter = ',', encoding = "ISO-8859-1")
csv
```

### DESKRIPSI DATA
Data Jumlah UMKM, Sumbangan PDB, Nilai Ekspor serta Tenaga Kerja yang diserap oleh UMKM tersebut serta dilengkapi dengan persentase pertumbuhan dari tahun ke tahun pada periode 1997-2012 di Indonesia


### STATISTIK DATA

```python
jumlahUMKM = csv[csv["indikator"]=="Jumlah UMKM"]
pertumbuhanUMKM = csv[csv["indikator"]=="Pertumbuhan Jumlah UMKM"]
jumlahTenagaKerja = csv[csv["indikator"]=="Jumlah Tenaga Kerja UMKM"]
pertumbuhanTenagaKerja = csv[csv["indikator"]=="Pertumbuhan Jumlah Tenaga Kerja UMKM"]
jumlahPDB = csv[csv["indikator"]=="Sumbangan PDB UMKM (harga konstan)"]
jumlahEkspor = csv[csv["indikator"]=="Nilai Ekspor UMKM"]

rataPertumbuhanUMKM = round(pertumbuhanUMKM.nilai.mean(), 2)
rataPertumbuhanTenagaKerja = round(pertumbuhanTenagaKerja.nilai.mean(), 2)
rataPDB = round(jumlahPDB.nilai.mean())
rataExpor = round(jumlahEkspor.nilai.mean())

print("Rata-Rata Pertumbuhan UMKM :", rataPertumbuhanUMKM, "% per tahun")
print("Rata-Rata Pertumbuhan Tenaga Kerja :", rataPertumbuhanTenagaKerja, "% per tahun")
print("Rata-Rata Expor : Rp.", rataExpor, " Milyar per tahun")
print("Rata-Rata Sumbangan PDB: Rp.", rataPDB, " Milyar per tahun")
```

### VISUALISASI SEDERHANA

Visualisasi perkembangan UMKM, Tenaga kerja, expor, dan sumbangan PDB

```python
jumlahUMKM.groupby("tahun")["nilai"].mean()
```

```python
graphUMKM = jumlahUMKM.groupby("tahun").mean()["nilai"]
graphUMKM = graphUMKM.plot(kind='bar', title ="Rata-Rata Jumlah UMKM per Tahun", legend=True, fontsize=12)
graphUMKM.set_xlabel("Tahun", fontsize=12)
graphUMKM.set_ylabel("Jumlah UMKM", fontsize=12)
```

```python
jumlahTenagaKerja.groupby("tahun")["nilai"].mean()
```

```python
graphTenagaKerja = jumlahTenagaKerja.groupby("tahun").mean()["nilai"]
graphTenagaKerja = graphTenagaKerja.plot(kind='bar', title ="Rata-Rata Jumlah Tenaga Kerja UMKM per Tahun", legend=True, fontsize=12)
graphTenagaKerja.set_xlabel("Tahun", fontsize=12)
graphTenagaKerja.set_ylabel("Jumlah UMKM", fontsize=12)
```

```python
jumlahPDB.groupby("tahun")["nilai"].mean()
```

```python
graphPDB = jumlahPDB.groupby("tahun").mean()["nilai"]
graphPDB = graphPDB.plot(kind='bar', title ="Rata-Rata Jumlah Sumbangan PDB UMKM per Tahun", legend=True, fontsize=12)
graphPDB.set_xlabel("Tahun", fontsize=12)
graphPDB.set_ylabel("Jumlah Sumbangan PDB", fontsize=12)
```

```python
jumlahEkspor.groupby("tahun")["nilai"].mean()
```

```python
graphEkspor = jumlahEkspor.groupby("tahun").mean()["nilai"]
graphEkspor = graphEkspor.plot(kind='bar', title ="Rata-Rata Jumlah Ekspor UMKM per Tahun", legend=True, fontsize=12)
graphEkspor.set_xlabel("Tahun", fontsize=12)
graphEkspor.set_ylabel("Jumlah Ekspor", fontsize=12)
```

### Narasi analisis

Jumlah UMKM mengalami penurunan pada tahun 1998 namun mengalami peningkatan kembali pada tahun 1999-2012. 

Penurunan tersebut berpengaruh pada jumlah tenaga kerja yang diserap dalam UMKM pada tahun 1998.

Dari sisi sumbangan PDB, penurunan jumlah UMKM tidak mempengaruhi besarnya Sumbangan PDB dari UMKM.

Dari sisi expor, penurunan jumlah UMKM tidak mempengaruhi besarnya nilai expor.

Dari data tersebut juga dapat disimpulkan bahwa:
- Rata-Rata Pertumbuhan UMKM di indonesia adalah 2.42 % per tahun
- Rata-Rata Pertumbuhan Tenaga Kerja yang diserap dalam UMKM di indonesia adalah 3.39 % per tahun
- Rata-Rata Expor UMKM di indonesia adalah Rp. 116472  Milyar per tahun
